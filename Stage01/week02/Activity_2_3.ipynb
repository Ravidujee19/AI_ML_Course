#Plot your own function using matplotlib

#Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

#Function Definition

```python
def fun(x):
  return x**2 * np.sin(x)
```

#X and Y Value Calculation

```python
x = np.linspace(-10, 10, 400)
y = fun(x)
```

#Plot

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x²·sin(x)', color='blue')
plt.title('Plot of f(x) = x²·sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
